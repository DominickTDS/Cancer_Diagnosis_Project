# Feature Selection Process

```python
from sklearn.model_selection import train_test_split
from feature_engine.selection import DropConstantFeatures
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_bc.drop(labels=['diagnosis'], axis=1), # drop the target
    df_bc['diagnosis'], # just the target
    test_size=0.3,
    random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

Removing Constant Features

```python
sel = DropConstantFeatures(tol=1, variables=None, missing_values='raise')

sel.fit(X_train)
```

```python
sel.features_to_drop_
```

Removing Correlated Features 

```python
from feature_engine.selection import DropCorrelatedFeatures, SmartCorrelatedSelection
```

```python

csel = DropCorrelatedFeatures(
    threshold=0.8,
    method='pearson',
    missing_values='ignore'
)
```

```python
csel.fit(X_train)
```

```python
csel.correlated_feature_sets_
```

```python
len(csel.features_to_drop_)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators=10,
    random_state=20,
    n_jobs=4,
)
```

```python
scsel = SmartCorrelatedSelection(
    variables=None, # if none, selector examines all numerical variables
    method="pearson",
    threshold=0.8,
    missing_values="raise",
    selection_method="model_performance",
    estimator=rf,
    scoring="roc_auc",
    cv=3,
)
```

```python
scsel.fit(X_train, y_train)
```

```python
scsel.correlated_feature_sets_
```

```python
len(scsel.features_to_drop_)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassificationReport
from yellowbrick.classifier import ROCAUC
```

```python
model = LogisticRegression()
```

```python
classes = ['B','M']
cm = ConfusionMatrix(model, classes=classes, percent=False)
cm.fit(X_train, y_train)
cm.score(X_test, y_test)
for label in cm.ax.texts:
    label.set_size(20)
cm.poof()
```

```python
# Instantiate the visualizer
visualizer = ClassificationReport(model, classes=classes)

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
g = visualizer.poof()
```
